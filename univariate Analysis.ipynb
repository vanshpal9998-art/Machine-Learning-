# *1. categorical Data*

*Categorical data is about grouping information into categories rather than measuring it numerically. It’s essential in surveys, research, and everyday classification tasks, especially when analyzing preferences, demographics, or rankings.*

***a.countplot***

*Definition: A countplot shows the frequency (count) of observations in each category of a categorical variable.

Purpose: It helps visualize how many times each category appears in your dataset.*

```python
sns.countplot(df['Survived'])
df['Survived'].value_counts()
```

```python
sns.countplot(df['Pclass'])
df['Pclass'].value_counts()
```

```python
sns.countplot(df['Sex'])
df['Sex'].value_counts()
```

```python
sns.countplot(df['Embarked'])
df['Embarked'].value_counts()
```

## *b. Piechart*

```python
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```
